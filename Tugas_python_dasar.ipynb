```python
"""
APLIKASI CONVERTER HARI DAN DETIK

input
Mau Menghitung apa? Hari atau Detik?

output:
361 hari = 1 tahun 0 bulan 1 hari

Ketentuan:
1 tahun = 360 hari
1 bulan = 30 hari
1 jam = 3600 detik
1 menit = 60 detik

"""
```

```python
print('*******APLIKASI CONVERTER HARI DAN DETIK*******\n')

x = input("Mau Menghitung apa? hari atau detik? ")
if x!='hari'and x!='detik': 
    print("masukkan hari atau detik")
else:
    print("Oke sekarang input berapa "+x+'nya')

y = int(input())
x = x.lower()

def hari(y):
    if y == 1:
        print(' hari adalah 24 jam')
    elif y % 365 == 0:
        y = int(y/365)
        print(' hari adalah',y,'tahun')
    elif y<365:
        tahun = 0
        bulan = int(y/30)
        hari = y%30
        print(y,' hari adalah %.f tahun %.f bulan %.f hari' %(tahun,bulan,hari))
    elif y>365:
        tahun = y/365
        bulan = int(y/30)-12
        hari = (y%365)
        print(y,' hari adalah %.f tahun %.f bulan %.f hari' %(tahun,bulan,hari))
    else:
        print('angka salah')
        return hari 

def detik(y):
    if y<=1:
        print('adalah 1 detik')
    elif y % 60 == 0:
        y = int(y/60)
        print(' detik adalah',y,'mnt')
    else:
        hr = y/3600 
        mnt = int(y/60) - 60
        sec = y%3600
        print(y, 'detik adalah %.f jam %.f menit %.f detik' %(hr, mnt, sec))
        return detik

if x == 'hari':
    hari(y)
else:
    detik(y)
```
