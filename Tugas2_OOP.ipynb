```python
"""
Afifah Kho'eriah

"""


#Class Parent bernama Game yang berisi method start_game dan print_result

class Game:
    def __init__(self,name,sex):
        #nilai akan diinputkan oleh user
        self.name = name
        self.sex = sex
        #inisiasi score 0 agar range nilai dari 1-5
        self.score = 0
        
    def start_game(self):
        #berisi 5 pertanyaan yang harus diisi dengan jawaban yes atau no
        self.question1 = input("Apakah kamu lebih suka warna putih daripada warna merah? ")
        self.question2 = input("Apakah kamu sangat cantik/tampan?  ")    
        self.question3 = input("Jika kamu sedang memotong daging ayam kamu lebih suka pakai pisau tumpul? ")
        self.question4 = input("Jika kamu liat ada korban kecelakaan kamu tidak akan menangis? ")
        self.question5 = input("Ketika kamu melukai seseorang sampai nangis apakah kamu akan membiarkannya? ")

        #perhitungan score dimana setiap jawaban yes akan mendapatkan nilai 1 
        self.score = 0
        user_choice = [self.question1,self.question2,self.question3,self.question4,self.question5]
        for choice in user_choice:
            if choice== 'yes':
                self.score +=1
    
    def print_result(self):
        print("Apakah kamu seorang psikopat? ")

#inheritance terhadap parent Game dan polymorphism terhadap method print_result
class female_game(Game):
    def print_result(self):
        print(
    '''
    INI GAME CEWE
    Test Result
    Name    : {}
    Sex     : {}
    Score   : {}
    '''.format(self.name,self.sex,self.score)
    )
        if self.score in range(3,6):
            print('Kamu seorang PSIKOPAT!!')
        else:
            print('Kamu BUKAN seorang PSIKOPAT')

#inheritance terhadap parent Game dan polymorphism terhadap method print_result
class male_game(Game):
    def print_result(self):
        print(
    '''
    INI GAME COWO
    Test Result
    Name    : {}
    Sex     : {}
    Score   : {}
    '''.format(self.name,self.sex,self.score)
    )
        if self.score in range(3,6):
            print('Kamu seorang PSIKOPAT!!')
        else:
            print('Kamu BUKAN seorang PSIKOPAT')

# ========================================================================================================


print('=======  Selamat Datang di Permainan Test Psikopat  =======\n')

name=input('siapa nama kamu? ')
sex =input('Apakah kamu cowo/cewe? ')
print("\n*****  Hallo",name," isi permainan ini dengan yes/no yaa  ***** \n")

# kondisi ketika user input gender
if sex.lower() == 'cowo':
    new_game = male_game(name,sex)
else:
    new_game = female_game(name,sex)
    
#print hasil    
new_game.start_game()
new_game.print_result()
```
